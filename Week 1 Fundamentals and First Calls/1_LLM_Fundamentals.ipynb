```python
import tiktoken

enc = tiktoken.get_encoding("cl100k_base")

sentences = [
    "Hello",
    "The transformer architecture uses multi-head attention.",
    "ChatGPT is an LLM.",
    "chat gpt is an llm.",
    "Tokenisation is a preprocessing step.",
    "I love pizza.",
    "The mitochondria is the powerhouse of the cell.",
    "import numpy as np",
    "2 + 2 = 4",
    "Supercalifragilisticexpialidocious",
]

for sentence in sentences:
    tokens = enc.encode(sentence)
    decoded = [enc.decode([t]) for t in tokens]
    print(f"Sentence: {sentence}")
    print(f"Token count: {len(tokens)}")
    print(f"Tokens: {decoded}")
    print()
```

```python
enc2 = tiktoken.get_encoding("p50k_base")

print(f"{'Sentence':<45} | cl100k | p50k")
print("-" * 65)
for sentence in sentences:
    count1 = len(enc.encode(sentence))
    count2 = len(enc2.encode(sentence))
    print(f"{sentence[:44]:<45} | {count1:<6} | {count2}")
```

```python
import os
import anthropic

os.environ["ANTHROPIC_API_KEY"] = "your-key-here"

api_key = os.environ.get("ANTHROPIC_API_KEY")
if not api_key:
    raise ValueError("ANTHROPIC_API_KEY is not set")

client = anthropic.Anthropic(api_key=api_key)
```

```python
response = client.messages.create(
    model="claude-sonnet-4-20250514",
    max_tokens=100,
    messages=[{"role": "user", "content": "Explain what a context window is in one sentence"}]
)

print("--- Full response object ---")
print(response)
print()
print("--- Just the text ---")
print(response.content[0].text)
```

```python
for i in range(3):
    response = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=100,
        messages=[{"role": "user", "content": "Explain what a context window is in one sentence"}]
    )
    print(f"Response {i+1}: {response.content[0].text}")
    print()
```
